## Task 1: Data Collection and Web Scraping.

- Description: Collect data from a website using web scraping techniques.


- Identify a target website and inspect its structure.

I am taking IPL information on Wikipedia. And this will  be scraped.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/Indian_Premier_League"
response = requests.get(url)
print(f" Its Response is {response} Which means the site allow scraping")
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Structure of the website
print(soup.prettify())
```

- Store the scraped data in a structured format (e.g., CSV, JSON).

```python
# Find all tables in alltables directory
if response.status_code == 200:
    tables = soup.find_all('table', {'class': 'wikitable'})
    for i, table in enumerate(tables):
        dfs = pd.read_html(str(table))
        if dfs:
            df = dfs[0]
            df.to_csv(f"alltables/table_{i}.csv", index=False)
            print(f"Table {i} saved as table_{i}.csv")
```
